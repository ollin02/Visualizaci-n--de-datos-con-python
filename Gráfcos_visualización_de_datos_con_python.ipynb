# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

```python
col_dict ={'Año':colombia.index.tolist(), 'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2015'])
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2015'])
plt.title('Inmigrantes de Colombia a Canada')
plt.xlabel('Años')
plt.ylabel('Inmigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes de Colombia a Canada')
ax.set_xlabel('Años')
ax.set_ylabel('Inmigrantes')
plt.show()
```
