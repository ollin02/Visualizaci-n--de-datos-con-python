# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

```python
col_dict ={'Año':colombia.index.tolist(), 'Inmigrantes':colombia.values.tolist()}
col_dict
```

## **Matplotlib**

https://matplotlib.org/

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2015'])
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2015'])
plt.title('Inmigrantes de Colombia a Canada')
plt.xlabel('Años')
plt.ylabel('Inmigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes de Colombia a Canada\n1980 -2013')
ax.set_xlabel('Años')
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,4))

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].set_title('Inmigrantes de Colombia a Canada\n1980 -2013')
axs[0].set_xlabel('Años')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la Inmigranción de Colombia a Canada\n1980 -2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()

plt.show()
```

```python
datos_col.describe()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Inmigración de los 4 mayores países sudamericanos \nhacia Canada 1980 -2013')

axs[0,0].plot(df.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

for ax in axs.flat:
  ax.set(xlabel='Años', ylabel='Número de Inmigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'],lw=2, marker='o', color='red')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes de Colombia a Canada', fontsize=18, loc='left')
ax.set_xlabel('Años', fontsize=14)
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')

plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)

colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
 'purple', 'brown', 'slateblue', 'gray', 'olive',
 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigrantes de sudamericanos hacia Canadí\nduramte el periodo de 1980 hasta2013 ',
             loc='left',fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.yaxis.set_tick_params(labelsize=10)
ax.xaxis.set_tick_params(labelsize=10)

plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='y', length=0)


ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudamérica: Colombia es el segundo país con mas inmigrantes\n hacia Canada duramte el periodo de 1980 hasta2013 ',
             loc='left',fontsize=18)
ax.set_ylabel('')
#ax.set_xlabel('Número de Inmigrantes', fontsize=14)
#ax.yaxis.set_tick_params(labelsize=10)
ax.xaxis.set_tick_params(labelsize=10)

for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j, i, str(j), va='center', fontsize=8.5, fontweight='bold', ha='left')

fig.savefig('/content/sample_data/inmigrantes_sudamericanos_canada.png', transparent=False, dpi=300, bbox_inches='tight')


plt.show()
```

## **Seaborn**

https://seaborn.pydata.org/tutorial/color_palettes.html

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

```python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index,orient='h')
ax.set(title='Top 10 países con mayor migración hacia Canada\n1980 - 2013',
       xlabel='Número de Inmigrantes', ylabel='año')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=top_10, x='Total', y=top_10.index,orient='h')
ax.set_title('Top 10 países con mayor migración hacia Canada\n1980 - 2013',
       loc='left', fontsize=16)
ax.set_xlabel('Número de Inmigrantes',fontsize=14)
ax.set_ylabel('')
plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))

  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canada\n1980 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')

  for i,j in enumerate(top_10['Total']):
    ax.text(j, i, str(j), va='center', fontsize=8.5, fontweight='bold', ha='left')

  plt.show()
```

```python
generar_grafico('Blues_r')
```

```python
generar_grafico('rocket')
```

```python
generar_grafico('coolwarm_r')
```

```python
generar_grafico('Paired')
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canada\n1980 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  sns.despine() #quita los vordes de los grafico

  for i,j in enumerate(top_10['Total']):
    ax.text(j, i, str(j), va='center', fontsize=8.5, fontweight='bold', ha='left')

  plt.show()
```

```python
sns.set_theme()
generar_grafico('tab10')
```

```python
sns.set_theme(style='whitegrid')
generar_grafico('tab10')
```

```python
sns.set_theme(style='white')
generar_grafico('tab10')
```

## **Plotly**

https://plotly.com/python/

```python
import plotly.express as px
```

```python
fig = px.line(datos_col, x='Año', y='Inmigrantes', title='Inmigrantes de Colombia a Canada')
fig.update_layout(width=1000,height=500,xaxis={'tickangle': -45},
                  xaxis_title='Año',yaxis_title='Número de Inmigrantes',
                  font_family='Arial',font_size=14, font_color='grey')
fig.show()
```

```python
fig = px.line(datos_col, x='Año', y='Inmigrantes',
              title='Inmigrantes de Colombia a Canada')
fig.update_traces(line_color='green', line_width=3)
fig.update_layout(width=1000,height=500,xaxis={'tickangle': -45},
                  xaxis_title='Año',yaxis_title='Número de Inmigrantes',
                  font_family='Arial',font_size=14, font_color='grey')
fig.show()
```
